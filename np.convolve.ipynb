# Numpy의 convolve 함수에 대해서 알아보자 

공식 : $(a * v)[n] = \sum_{m = -\infty}^{\infty} a[m] v[n - m]$

np.convolve(input, weight, mode)

- input을 smoothing하는 방법으로 mode에 따라 가중 평균을 어떻게 적용할지 달라진다. 

- mode가 full인 경우 나오는 output은 m+n-1의 벡터
- mode가 same일 때는 input의 차원이나 weight의 차원 중 큰 값을 기준으로 나온다.(MAX(M,N))
- mode가 valid일 경우에는 MAX(M,N)-MIN(M,N)+1의 벡터가 나온다.

- 식에 대한 이해가 완료

```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# 사용법 
input_ = np.array([2,1,5,2,4,3])
weight = np.array([0,1,0.5])

y = np.convolve(input_,weight,mode = 'full')
y1 = np.convolve(input_,weight,mode = 'same')
y2 = np.convolve(input_,weight,mode='valid')


plt.plot(input_,label ="original")
plt.plot(y,label='full convolution')
plt.plot(y1,label="same convolution")
plt.plot(y2,label="valic convolution")
plt.legend()
```
