# 데이터 시각화 

- 숫자를 그림으로 바꾸는 작업 

데이터 시각화는 데이터 분석에 있어서 아래의 3가지를 가능하게 합니다. 
- 데이터에 대한 이해도 증진 
- 데이터를 통한 새로운 특징 탐색 
- 데이터를 통한 문제점 도출 (현재 데이터 수집에 문제가 없는지?) : 굉장히 중요한 작업 
- 데이터의 전체와 부분들을 동시에 이해가 가능 
- 가설에 대한 수립과 검증이 가능

수치 분석만으로는 데이터 파악이 불충분합니다.(데이터 시각화가 필요한 이유) 
- 단순 fitting으로는 의미가 데이터 전체에 대한 파악이 불가능하다.

데이터 시각화의 조건 

- 예뻐야 합니다. 
- 군더더기가 적어야 합니다. (메세지에 대한 전달을 잘 해줘야 합니다) 
- 그림을 정확하게 그려야 합니다.

# 대학원 이후의 삶에서 겪는 일 
- 같은 그림을 여기저기 사용   
ex) 랩미팅, 학회, 논문, 특허, 보고, ...  
ex) 상사의 취향에 맞도록 그림을 계속 수정: 폰트,색감, ...

- 같은 그림에 데이터만 갈아끼우기 
ex) 주간보고, 현황판 

- 유료 프로그램 사용 불가 
ex)OriginLab



# Code 기반 시각화 장점

- 여러 그림을 동일하게 수정하기 편리함   
    ex) 파랑색을 초록으로 바꿔야 되는 그림 100장

- 그림에 안 보이는 주석 달기 가능   
    ex) "00부분 강조","전무님 관심사항" 

- 예전 작업 ctrl+c/v 유리   
ex) 엑셀로 마우스 클릭 100번 vs 예전 코드 3개 불러와서 짜집기 

- 재현성   
ex) 마우스 클릭해서 y=0에 정확하게 줄긋기 vs 줄긋기 명령에 y=0 입력하기
